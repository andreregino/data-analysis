## Step 2: Data Preparation
- Drop columns and rows
- Identify duplicate rows
- Rename columns
- Feature engineer

```python
## "speed" is an object and speed_mph is a numeric version of it
df = df[['coaster_name', 'Location', 'Status',
       'Manufacturer', 'Type_Main',
       'opening_date_clean',
       'year_introduced',
       'speed_mph', 'height_ft',
       'Gforce_clean']].copy()
```

```python
# the 57 columns turned into 10
df.head(3)
```

```python
df.dtypes
```

```python
# transforming date clean (object) into datetime column
df['opening_date_clean'] = pd.to_datetime(df['opening_date_clean'])
```

```python
# Renaming Columns
df = df.rename(columns={"Location":"location",
                   "Status": "status",
                   "Manufacturer": "manufacturer",
                   "Type_Main": "type",
                   "opening_date_clean": "opening_date",
                   "Gforce_clean": "g_force"
})
```

```python
# missing values
df.isna().sum()
```

```python
# verify if there is a whole line duplication
df.loc[df.duplicated()]
```

```python
# locate the duplicate by name of the rollercoaster... might be an insertion error
df.loc[df.duplicated(subset=["coaster_name"])]
```

```python
# examining a duplicate coaster name example. The two lines are equal, except from the year introduced
df.query("coaster_name == 'Crystal Beach Cyclone'")
```

```python
# getting all values that are NOT (look at the ~ sign) duplicate based on these columns
# and update the current df
# the index got non-consecutive, since we remove some lines. Reset index and drop=True fix it

df = df.loc[~df.duplicated(subset=["coaster_name", "location", "opening_date"])] \
    .reset_index(drop=True) .copy()
```

```python
df.head(3)
```

## Step 3: Feature Understanding
or, in other words, univariate analysis
- plotting feature distribution (histogram, KDE and boxplot)

```python
# well, from 98 to 2002, we had a "boom" in rollercoaster introduction
df['year_introduced'].value_counts().head(10)
```

```python
# lets plot it with matplot
ax = df['year_introduced'].value_counts() \
                    .head(10) \
                    .plot(kind="bar", title="Top 10 Years Coasters Introduced")

ax.set_xlabel("Year Introduced")
ax.set_ylabel("Count")
```

```python
# it seems that most of the coasters reaches 50 mph. let's see it changing the plot to kde
ax = df["speed_mph"].plot(kind="kde", title="Coaster speed (mph)")
ax.set_xlabel("Speed (mph)")
```

```python
# what about the g force?
df['g_force'].value_counts().head(10)
```

```python
ax = df['g_force'].value_counts() \
                    .head(10) \
                    .plot(kind="bar", title="Top 10 G-force")

ax.set_xlabel("G-force")
ax.set_ylabel("Count")
```

## Step 4: Feature Relationships
in other words, multivariable analysis
- Scatterplot
- Heatmap 
- Pairplot
- Groupby comparisons

```python
# Comparing the speed and height
# it seems the highest, the faster..
df.plot(kind="scatter", 
       x = "speed_mph",
       y = "height_ft",
       title = "Coaster speed vs height")
plt.show()
```

```python
# advancing using seaborn plots...
# is there a correlation between speed, height and g force?
sns.scatterplot(x = "speed_mph",
                y = "height_ft",
                data = df,
                hue="g_force")
```

```python
# comparing each pair of our 4 numeric values to verify if there is a relation
# also, we use hue = type to verify if there is a pattern based on the type
sns.pairplot(df, vars = ["year_introduced", "speed_mph", "height_ft", "g_force"],
                 hue="type")
```

What can we see there:
- The years_introduced rises and the type gradually changes to steel
- The fasters and highers are the steel type

```python
# lets see the correlations
df_corr = df[["year_introduced", "speed_mph", "height_ft", "g_force"]].dropna().corr()
df_corr
```

```python
# see in heatmap. annot=True shows the values inside the squares
sns.heatmap(df_corr)
```

What we can see there:
- year is not correlated with anything
- speed in very correlated with height and g_force

## Step 5: Asking questions about the data

What are the locations with faster rollercoasters (minimum of 10)?

```python
#group by location and return the value of speed only
ax = df.query("location != 'Other'") \
    .groupby("location")["speed_mph"] \
    .agg(["mean", "count"]) \
    .query("count >= 10") \
    .sort_values("mean")["mean"]
ax
```

```python
ax.plot(kind="barh",
       figsize=(12,5), title="Average Coast speed by Location").set_xlabel("Average Coaster Speed")

plt.show()
```
